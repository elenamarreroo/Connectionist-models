En este primer ejemplo se ha utilizado un generador de funciones de pertenencia triangular. Ahora, lo que se va a hacer repetir el ejercicio utilizando la función del trapezoide. Esta es muy parecida a la triangular pero es un buen ejemplo para entender como funcionan los parámetros. 

La función del trapezoide viene dada por skfuzzy.trapmf(x, abcd). Se diferencia de la primera en que tenemos un punto más de referencia. En nuestro caso, vamos a probar por un lado los valores [0, 0, 2, 5], considerando estos los números cercanos a cero y por otro lado los valores [2, 4.5, 6.5,9] para los valores cercanos a 5.


````python
# 2. Rango de valores donde se analizan los conjuntos y definición de ellos.
reales= np.arange(0, 11,0.1) 
A= sk.trapmf(reales, [0, 0, 2, 5]) 
B= sk.trapmf(reales, [2, 4.5, 6.5,9])

# 3. Representamos los conjuntos.
plt.figure()
plt.plot(reales, A, 'b', linewidth=1.5, label='Conjunto A') 
plt.plot(reales, B, 'k', linewidth=1.5, label='Conjunto A')
````

````python
# 4. Realizamos las operaciones AND/OR y complemento.
intersec=sk.fuzzy_and(reales,A,reales,B)[1] 
union=sk.fuzzy_or(reales,A,reales,B)[1] 
CA=sk.fuzzy_not(A)
CB=sk.fuzzy_not(B)

# 5. Representamos los resultados.
plt.figure()
plt.plot(reales, intersec, 'b', linewidth=1.5,label='Interseccion') 
plt.plot(reales, union, 'k', linewidth=1.5, label='Union')
plt.figure()
plt.plot(reales, CA, 'r', linewidth=1.5, label='Complemento A') 
plt.plot(reales, CB, 'm', linewidth=1.5, label='Complemento B')
````

Ahora hemos visto dos funciones de preferencia que son muy parecidas, vamos a probar la función de pertenencia difusa gaussiana ya que esta tiene una estructura diferente.

La función de la gaussiana viene dada por skfuzzy.gaussmf(x, media, sigma). Siendo la media el parámetro gaussiano para el valor central (medio) y sigma el parámetro gaussiano para la desviación estándar. Sabiendo esto, podemos definir para A la media = 0 y una sigma = 1 para que no se nos agrande mucho la función. A su vez podemos decir que los valores de B podrían ser media = 5 y sigma = 1.


````python
# 2. Rango de valores donde se analizan los conjuntos y definición de ellos.
reales= np.arange(0, 11,0.1) 
A= sk.gaussmf(reales, 0, 1) 
B= sk.gaussmf(reales, 5, 1)

# 3. Representamos los conjuntos.
plt.figure()
plt.plot(reales, A, 'b', linewidth=1.5, label='Conjunto A') 
plt.plot(reales, B, 'k', linewidth=1.5, label='Conjunto B')
````

````python
# 4. Realizamos las operaciones AND/OR y complemento.
intersec=sk.fuzzy_and(reales,A,reales,B)[1] 
union=sk.fuzzy_or(reales,A,reales,B)[1] 
CA=sk.fuzzy_not(A)
CB=sk.fuzzy_not(B)

# 5. Representamos los resultados.
plt.figure()
plt.plot(reales, intersec, 'b', linewidth=1.5,label='Interseccion') 
plt.plot(reales, union, 'k', linewidth=1.5, label='Union')
plt.figure()
plt.plot(reales, CA, 'r', linewidth=1.5, label='Complemento A') 
plt.plot(reales, CB, 'm', linewidth=1.5, label='Complemento B')
````

````python
# 1. Cargamos las librerías.
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 
import matplotlib.pyplot as plt

# 2. Definimos la variable de entrada y la visualizamos
ilumina = ctrl.Antecedent(np.arange(0, 10.1, 0.1), 'ilumina')
ilumina['baja'] = fuzz.sigmf(ilumina.universe, 2, -7) 
ilumina['media'] = fuzz.trapmf(ilumina.universe, [2, 4, 6, 8]) 
ilumina['alta'] = fuzz.trapmf(ilumina.universe, [6, 7, 10,11]) 
ilumina.view()
````

````python
# 3. Definimos la variable de salida y lo visualizamos
pot = ctrl.Consequent(np.arange(0,101,1), 'Potencia') 
pot['baja'] = fuzz.trimf(pot.universe, [0, 0, 25]) 
pot['media'] = fuzz.trimf(pot.universe, [25, 50, 75]) 
pot['alta'] = fuzz.pimf(pot.universe, 50,80,100,125) 
pot.view()
````

````python
# 4. Definimos las reglas
regla1 = ctrl.Rule(ilumina['baja'],pot['alta'])
regla2 = ctrl.Rule(ilumina['media'],pot['media'])
regla3 = ctrl.Rule(ilumina['alta'], pot['baja'])

# 5. Definimos el sistema de inferencia.
ilum_ctrl = ctrl.ControlSystem([regla1, regla2, regla3])
ilum_sim = ctrl.ControlSystemSimulation(ilum_ctrl)

# 6. Simulamos el sistema y visualizamos la salida.
ilum_sim.input['ilumina'] = 6.5
ilum_sim.compute()
ilum_sim.output
pot.view(sim=ilum_sim)
````

-  Desarrolla un sistema de Mamdani cuya salida sea la velocidad de cruce de un peatón en una calle sin paso de cebra y cuyas variables de entrada sean velocidad del coche que se acerca (entre 0 y 80 kms/h) y espacio que se tiene (entre 0 y 200 metros). Lo único a conocer con respecto al ejemplo anterior es que cuando tienes que combinar reglas lo haces de la siguiente forma: R1 = ctrl.Rule(V1[‘setA'] & V2[‘setB’],V3[‘setC’]).Si usas el operador or en vez de & usas |.

````python
# 1. Cargamos las librerías.
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# 2.1. Definimos la variable de entrada distancia y la visualizamos 
distancia = ctrl.Antecedent(np.arange(0, 200, 1), 'Distancia')
distancia['bajo'] =  fuzz.trimf(distancia.universe, [0, 0, 75])
distancia['medio'] = fuzz.gbellmf(distancia.universe, 50, 10, 112.5)
distancia['alto']  = fuzz.trimf(distancia.universe, [150, 200, 200])
distancia.view()
````

````python
# 2.2. Definimos la variable de entrada velocidad y la visualizamos 
velocidad = ctrl.Antecedent(np.arange(0, 80, 1), 'Velocidad')
velocidad['bajo'] = fuzz.trimf(velocidad.universe, [0, 0, 20])
velocidad['medio'] = fuzz.gbellmf(velocidad.universe, 8, 6, 25)
velocidad['alto']  = fuzz.trimf(velocidad.universe, [35, 80, 80])
velocidad.view()
````

````python
# 3. Definimos la variable de salida y lo visualizamos
peaton = ctrl.Consequent(np.arange(0,16,1), 'Peaton') 
peaton['bajo'] = fuzz.trimf(peaton.universe, [0, 0, 3]) 
peaton['medio'] = fuzz.trapmf(peaton.universe, [2, 5, 8, 10]) 
peaton['alto'] = fuzz.pimf(peaton.universe, 8,14,15,16) 
peaton.view()
````

````python
# 4. Definimos las reglas
R1 = ctrl.Rule(velocidad['bajo'] & distancia['alto'],peaton['bajo'])

R2 = ctrl.Rule((velocidad['alto'] & distancia['alto'])|(velocidad['alto'] & distancia['medio'])|
                   (velocidad['bajo'] & distancia['bajo'])|(velocidad['bajo'] & distancia['medio'])|
                   (velocidad['medio']& distancia['medio'])|(velocidad['medio'] & distancia['bajo'])|(velocidad['medio'] & distancia['alto'])
                   , peaton['medio'])

R3 = ctrl.Rule(velocidad['alto'] & distancia['bajo'],peaton['alto'])
````

````python
# 5. Definimos el sistema de inferencia.
cruz_ctrl = ctrl.ControlSystem([R1, R2, R3])
cruz_sim = ctrl.ControlSystemSimulation(cruz_ctrl)

# 6. Simulamos el sistema y visualizamos la salida.
cruz_sim.input['Distancia'] = 25
cruz_sim.input['Velocidad'] = 70
cruz_sim.compute()
cruz_sim.output
peaton.view(sim=cruz_sim)
````
