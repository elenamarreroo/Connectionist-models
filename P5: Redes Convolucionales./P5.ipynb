````python
### 1. Carga de librerías.
import keras
from keras.models import Sequential
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from keras.layers import Dense, Dropout, Flatten, Input 
from keras.layers import Conv2D, MaxPooling2D

### 2. Carga de los datos.
mnist = tf.keras.datasets.mnist
(img_rows, img_cols) = (28,28)
(x_train, y_train),(x_test, y_test) = mnist.load_data()

### 3. Codificación/normalización de los datos.
# Como solo tienen que tener dotos en 3d configuramos las dimensiones para trabajar con blanco y negro. Si no hacemos esto, nos daria fallo de dimensiones
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
x_train, x_test = x_train / 255.0, x_test / 255.0 
x_train_cnn = np.expand_dims(x_train, -1) 
x_test_cnn = np.expand_dims(x_test, -1)

### 4. Creamos el modelo.
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1), padding='same'))
model.add(MaxPooling2D(pool_size=(4, 4)))
model.add(Flatten())
model.add(Dense(10, activation='softmax'))

### 5. Compilamos el modelo.
model.compile(loss= 'categorical_crossentropy', optimizer= 'adam', metrics=[ 'accuracy' ])

### 6. Obtenemos información del modelo.
model.summary()

### 7. Ajustamos el modelo.
ajuste=model.fit(x_train_cnn, y_train, epochs=20)

### 8. Evaluamos el modelo.
medida = model.evaluate(x_test_cnn, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], medida[1]*100))
````

````python
### 1. Carga de librerías.
import keras
from keras.models import Sequential
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from keras.layers import Dense, Dropout, Flatten, Input 
from keras.layers import Conv2D, MaxPooling2D

### 2. Carga de los datos.
mnist = tf.keras.datasets.mnist
(img_rows, img_cols) = (28,28)
(x_train, y_train),(x_test, y_test) = mnist.load_data()

### 3. Codificación/normalización de los datos.
# Como solo tienen que tener dotos en 3d configuramos las dimensiones para trabajar con blanco y negro. Si no hacemos esto, nos daria fallo de dimensiones
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
x_train, x_test = x_train / 255.0, x_test / 255.0 
x_train_cnn = np.expand_dims(x_train, -1) 
x_test_cnn = np.expand_dims(x_test, -1)
````

````python
### 4. Creamos el modelo.
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1), padding='same'))
model.add(MaxPooling2D(pool_size=(4, 4)))
model.add(Flatten())
model.add(Dense(20))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

### 5. Compilamos el modelo.
model.compile(loss= 'categorical_crossentropy', optimizer= 'rmsprop', metrics=[ 'accuracy' ])

### 6. Obtenemos información del modelo.
model.summary()

### 7. Ajustamos el modelo.
ajuste=model.fit(x_train_cnn, y_train, epochs=20)

### 8. Evaluamos el modelo.
medida = model.evaluate(x_test_cnn, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], medida[1]*100))
````

**• Añadiendo otra capa de neuronas ocultas antes de la capa de salida. Introduce dropout (considera p=0.5).**

La capa de abandono establece aleatoriamente las unidades de entrada en 0 con una frecuencia de rate en cada paso durante el tiempo de entrenamiento, lo que ayuda a evitar el sobreajuste. Las entradas que no están configuradas en 0 se escalan en 1 / (1 - tasa) de modo que la suma de todas las entradas no cambia.

Nosotras hemos probado y comparado sin dropout y con dropout con el parámetro p = 0.5.

1. Sin dropout obtenemos un *accuracy: 98.47%%*

2. Con p = 0.2 obtenemos un *accuracy: 98.65%.*

3. Con p = 0.5 obtenemos un *accuracy: 98.66%.*

Por lo que podemos decir que en este caso ha mejorado el funcionamiento añadiendo los términos de dropout con un p = 0.5, cuanto mayor sea el valor más rápido se completa el proceso.

4. Si añadimos una capa de 500 neuronas combinandolo con un Dropout de p=0.5 obtenemos un *accuracy:98.71%*

5. Si añadimos una capa de 100 neuronas combinandolo con un Dropout de p=0.5 obtenemos un *accuracy: 98.85%*

5. Si añadimos una capa de 1000 neuronas combinandolo con un Dropout de p=0.5 obtenemos un *accuracy: 98.40%*

Por lo tanto, las conclusiones que podemos obtener es que a menor número de neuronas, mayor número de capas y con un dropout (p = 0.5) mejora el funcionamiento del modelo. El mejor de los casos ha sido el punto 5.

**• Algoritmos de optimización (por ejemplo el RMSprop).** 

Un optimizador es uno de los dos argumentos necesarios para compilar un modelo de Keras. Se puede crear una instancia de un optimizador antes de pasarlo a model.compile(), o puede pasarlo por su identificador de cadena.

Podemos probar con los algoritmos de optimización siguientes: SGD, RMSprop, Adán, Adadelta, Adagrad, Adamax, Nadam y Ftrl

Nosotras hemos probado y comparado el optimizer= 'adam',  el optimizer= 'rmsprop', el optimizer = 'Adadelta' y el optimizer = 'SGD'.

1. El optimizer = 'adam' es un método de descenso de gradiente estocástico que se basa en la estimación adaptativa de momentos de primer y segundo orden. Hemos obtenido un *accuracy: 98.47%%*

2. El optimizer= 'rmsprop' es, por un lado antener un promedio móvil (descontado) del cuadrado de gradientes y a su vez, divida el gradiente por la raíz de este promedio. Esta implementación de RMSprop utiliza un impulso simple, no el impulso de Nesterov. Hemos obtenido un *accuracy: 98.41%*

3. El optimizer= 'Adadelta' es un método de descenso de gradiente estocástico que se basa en la tasa de aprendizaje adaptativo por dimensión. Hemos obtenido un *accuracy: 76.43%*

3. El optimizer= 'SGD' es un método de descenso de gradientes (con impulso). Hemos obtenido un *accuracy: 97.72%*

En este caso, no ha mejorado el funcionamiento con el optimizer= 'rmsprop' ni con el optimizer= 'Adadelta' ni con optimizer= 'SGD'. El optimizador que tiene el mejor funcionamiento es optimizer = 'adam'.
